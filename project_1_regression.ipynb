#Dealing with missing values

```python
nas = df_train.isna().sum()
nas[nas>0]
```

```python
# replace the missing values with the mean

## for convenience, create a function to do that:
def fill_na_mean(df, column):
    df[column] = df[column].fillna(df[column].mean())

## define a list of columns to be imputed
columns_to_mean = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']

## apply the function to the dataframe
for column in columns_to_mean:
    fill_na_mean(df_train, column)

## Note: this is the primitive syntax that was replaced by the function above and the for loop above:
# df_train['LotFrontage'] = df_train['LotFrontage'].fillna(df_train['LotFrontage'].mean())
# df_train['MasVnrArea'] = df_train['MasVnrArea'].fillna(df_train['MasVnrArea'].mean())
# df_train['GarageYrBlt'] = df_train['GarageYrBlt'].fillna(df_train['GarageYrBlt'].mean())
```

```python
print(df_train.shape)
nas1 = df_train.isna().sum()
print(nas1[nas1>0])
```

```python
# replace the missing values with the most frequent value (see the replacement with the mean)

## for convenience, create a function to do that:
def fill_na_mode(df, column):
    df[column] = df[column].fillna(df[column].mode()[0])

## define a list of columns to be imputed
columns_to_mode = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                   'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']

## apply the function to the dataframe
for column in columns_to_mode:
    fill_na_mode(df_train, column)
```

```python
print(df_train.shape)
nas2 = df_train.isna().sum()
print(nas2[nas2>0])
```

#Polynomial features

#Polynomial features - universal

```python
# with splitting the data into train and test
# with scaling the data

from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn import config_context


for i in range(1, 3):
  poly = PolynomialFeatures(i, include_bias=True)
  with config_context(transform_output="pandas"):
      # the output of transform will be a Pandas DataFrame
      scaler = StandardScaler()
      X_train_poly = poly.fit_transform(X_train)
      X_train_scaled = scaler.fit_transform(X_train_poly)
      X_test_poly = poly.transform(X_test)
      X_test_scaled = scaler.transform(X_test_poly)

  reg_poly = LinearRegression()
  reg_poly.fit(X_train_scaled, y_train)


  print(f'All features with Polynomial features with degree {i}, with Y-intercept - train', reg_poly.score(X_train_scaled, y_train))
  print(f'All features with Polynomial features with degree {i}, with Y-intercept - test', reg_poly.score(X_test_scaled, y_test))
  print(X_train_scaled.shape)
```

```python
# without scaling the data
for i in range(1, 3):
  poly = PolynomialFeatures(i, include_bias=True)
  with config_context(transform_output="pandas"):
      # the output of transform will be a Pandas DataFrame
      X_train_poly = poly.fit_transform(X_train)
      X_test_poly = poly.transform(X_test)

  reg_poly = LinearRegression()
  reg_poly.fit(X_train_poly, y_train)

  print(f'All features with Polynomial features with degree {i}, with Y-intercept - train', reg_poly.score(X_train_poly, y_train))
  print(f'All features with Polynomial features with degree {i}, with Y-intercept - test', reg_poly.score(X_test_poly, y_test))
  print(X_train_poly.shape)
```
